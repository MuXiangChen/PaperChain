# blueprint
blueprint是文章的核心结构框架，ideagraph和outlineweaving接受blueprint作为约束，对用户输入的内容进行整理和编排，blueprint应该由ss和ca自动生成，但是也可以用户手写创建


```python
# ===== Blueprint 模块测试：环境初始化 =====

import sys
import importlib
from pathlib import Path

project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

import src.models.blueprint as bp
bp = importlib.reload(bp)

Blueprint = bp.Blueprint
BlueprintSection = bp.BlueprintSection
BlueprintSlot = bp.BlueprintSlot
WritingGuideline = bp.WritingGuideline
HardConstraint = bp.HardConstraint
SelfCheckRule = bp.SelfCheckRule

```

```python
# ===== 手动构建 MBA Blueprint =====

def build_mba_blueprint() -> Blueprint:
    blueprint = Blueprint(
        blueprint_id="bp_mba_001",
        name="MBA 论文写作 Blueprint",
        paper_type="MBA Thesis",
        discipline="Business Administration",
        source="Standards System + Canon Analyzer",
        description="用于 MBA 论文写作的强约束模板。"
    )

    # -------------------------
    # 摘要
    # -------------------------
    abstract_guideline = WritingGuideline(
        notes=["摘要应高度概括全文，不展开细节。"]
    )
    abstract_guideline.add_constraint(HardConstraint(
        constraint_type="max_sentences",
        description="摘要应保持简洁，避免过长。",
        value=8
    ))
    abstract_guideline.add_self_check(SelfCheckRule(
        check_type="coverage_check",
        description="检查摘要是否覆盖背景、方法、结论。",
        criteria="must_cover_background_method_conclusion"
    ))

    abstract = BlueprintSection(
        section_id="abstract",
        title="摘要",
        purpose="概述研究问题、研究方法、核心结论和管理启示。",
        description="摘要用于高度概括全文，不展开细节。",
        order=1,
        guideline=abstract_guideline
    )

    abstract.add_slot(BlueprintSlot(
        slot_id="abstract_1",
        name="研究背景与问题",
        purpose="简要交代研究背景并说明核心问题。",
        expected_content="说明研究场景、问题来源以及研究对象。",
        material_types=["background", "problem_statement"],
        required=True,
        citations_required=False,
        order=1
    ))

    abstract.add_slot(BlueprintSlot(
        slot_id="abstract_2",
        name="研究方法",
        purpose="概述论文采用的研究方法或分析路径。",
        expected_content="简要说明所使用的方法、框架或案例分析方式。",
        material_types=["method", "method_detail"],
        required=True,
        citations_required=False,
        order=2
    ))

    abstract.add_slot(BlueprintSlot(
        slot_id="abstract_3",
        name="研究结论与启示",
        purpose="概述研究结论和管理启示。",
        expected_content="简要给出核心研究发现及其实践意义。",
        material_types=["result", "conclusion", "management_implication"],
        required=True,
        citations_required=False,
        order=3
    ))

    # -------------------------
    # 第一章：研究背景与问题提出
    # -------------------------
    section1_guideline = WritingGuideline(
        notes=["本章负责建立研究语境，并明确论文核心问题。"]
    )
    section1_guideline.add_constraint(HardConstraint(
        constraint_type="min_citations",
        description="本章至少应包含 3 处引用。",
        value=3
    ))
    section1_guideline.add_self_check(SelfCheckRule(
        check_type="logic_check",
        description="检查本章是否遵循 背景 → 问题 → 动机/目标 的逻辑顺序。",
        criteria="background_then_problem_then_motivation"
    ))

    section1 = BlueprintSection(
        section_id="1",
        title="研究背景与问题提出",
        purpose="建立研究背景，提出问题，说明研究动机与目标。",
        description="本章负责把读者带入研究情境，并明确本文要解决的问题。",
        order=2,
        guideline=section1_guideline
    )

    section1.add_slot(BlueprintSlot(
        slot_id="1_1",
        name="研究背景",
        purpose="说明现实背景、行业背景或管理实践背景。",
        expected_content="介绍研究所依托的现实问题、行业场景或管理挑战。",
        material_types=["background", "evidence", "example"],
        required=True,
        citations_required=True,
        order=1
    ))

    section1.add_slot(BlueprintSlot(
        slot_id="1_2",
        name="研究问题",
        purpose="明确提出本文研究的核心问题。",
        expected_content="从背景中抽取并明确界定研究问题。",
        material_types=["problem_statement", "definition"],
        required=True,
        citations_required=False,
        order=2
    ))

    section1.add_slot(BlueprintSlot(
        slot_id="1_3",
        name="研究动机与目标",
        purpose="说明为什么值得研究以及论文目标是什么。",
        expected_content="解释研究必要性，并明确本文想解决什么。",
        material_types=["motivation", "objective"],
        required=True,
        citations_required=False,
        order=3
    ))

    # -------------------------
    # 第二章：国内外研究现状
    # -------------------------
    section2_guideline = WritingGuideline(
        notes=["本章重点是文献综述与研究评述，必须体现研究定位。"]
    )
    section2_guideline.add_constraint(HardConstraint(
        constraint_type="min_citations",
        description="文献综述部分至少应包含 8 处引用。",
        value=8
    ))
    section2_guideline.add_self_check(SelfCheckRule(
        check_type="citation_check",
        description="检查本章是否有充分引用支撑。",
        criteria="citations_sufficient_for_literature_review"
    ))
    section2_guideline.add_self_check(SelfCheckRule(
        check_type="logic_check",
        description="检查是否先综述再评述。",
        criteria="review_then_commentary"
    ))

    section2 = BlueprintSection(
        section_id="2",
        title="国内外研究现状",
        purpose="回顾现有研究，建立论文研究定位。",
        description="本章对国内外相关文献进行梳理、分类、比较与评述。",
        order=3,
        guideline=section2_guideline
    )

    section2.add_slot(BlueprintSlot(
        slot_id="2_1",
        name="国外研究现状",
        purpose="整理国外已有研究。",
        expected_content="归纳国外研究方向、方法和主要结论。",
        material_types=["related_work", "comparison", "evidence"],
        required=True,
        citations_required=True,
        order=1
    ))

    section2.add_slot(BlueprintSlot(
        slot_id="2_2",
        name="国内研究现状",
        purpose="整理国内已有研究。",
        expected_content="归纳国内研究重点，并与国外研究进行对照。",
        material_types=["related_work", "comparison", "evidence"],
        required=True,
        citations_required=True,
        order=2
    ))

    section2.add_slot(BlueprintSlot(
        slot_id="2_3",
        name="研究评述",
        purpose="总结现有研究的不足，并引出本文研究切入点。",
        expected_content="概括已有研究不足、研究空白与本文定位。",
        material_types=["analysis", "discussion", "problem_statement"],
        required=True,
        citations_required=True,
        order=3
    ))

    blueprint.add_section(abstract)
    blueprint.add_section(section1)
    blueprint.add_section(section2)

    return blueprint

mba_blueprint = build_mba_blueprint()
mba_blueprint.print_tree()
```

# blueprint保存和加载
保存到本地并加载blueprint

```python
BlueprintCompiler = bp.BlueprintCompiler
TreePrinter = bp.TreePrinter
compiler = BlueprintCompiler()

import json


# 1️⃣ Blueprint -> dict
data = compiler.to_dict(blueprint)

# 2️⃣ dict -> JSON 保存到本地
file_path = "blueprint_test.json"

with open(file_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("Blueprint 已保存到:", file_path)

# 3️⃣ 从 JSON 文件读取
with open(file_path, "r", encoding="utf-8") as f:
    spec = json.load(f)

# 4️⃣ dict -> Blueprint
blueprint_loaded = compiler.from_dict(spec)

# 5️⃣ 打印
print("\n加载后的 Blueprint 对象:")

tree_printer = TreePrinter()
tree_printer.print_blueprint(blueprint_loaded)

```
