```python
import sys
import os

sys.path.append(os.path.abspath(".."))

import import_ipynb

from research.Canon_Analyzer import canon_analyzer
from research.Standards_System import standards_system
from research.IdeaGraph_Engine import ideagraph_engine
from research.Idea_Workspace import idea_workspace
from research.Outline_Weaving_Engine import outline_weaving
from research.Editing_Refinement_Engine import editing_refinement
from research.Assist_Plugins import assist_plugins
from research.FigureLab import figurelab
from research.Export_Renderers import export_renderers
```

```python

# 1. Standards System
standards = standards_system.StandardsSystem()
pack = standards.get_pack("cs")
print("=== Standards Pack ===")
print(pack)

# 2. Canon Analyzer
canon = canon_analyzer.CanonAnalyzer()
guideline = canon.analyze(["sample_paper_1.pdf", "sample_paper_2.pdf"])
print("\n=== Guideline ===")
print(guideline)

# 3. IdeaGraph Engine
ideagraph = ideagraph_engine.IdeaGraphEngine()
graph = ideagraph.build_graph([
    "I want to write about transformer-based academic writing systems.",
    "I also want to discuss retrieval and document export.",
])
print("\n=== Idea Graph ===")
print(graph)

# 4. Idea Workspace & Evidence Builder
evidence_builder = idea_workspace.IdeaWorkspaceEvidenceBuilder()
workspace = evidence_builder.build_workspace(graph, pack["retrieval_sources"])
print("\n=== Evidence Workspace ===")
print(workspace)

# 5. Outline Weaving Engine
weaver = outline_weaving.OutlineWeavingEngine()
woven = weaver.weave(guideline, graph, workspace)
print("\n=== Woven Outline ===")
print(woven)

# 6. Editing & Refinement Engine
refinement = editing_refinement.EditingRefinementEngine()
paper_ir = refinement.refine(woven, guideline, pack["citation_style"])
print("\n=== Paper IR ===")
print(paper_ir)

# 7. Assist Plugins
plugins = assist_plugins.AssistPlugins()
checked = plugins.grammar_check("This is a demo paragraph.")
print("\n=== Grammar Check ===")
print(checked)

# 8. FigureLab
figurelab_instance = figurelab.FigureLab()
figure_spec = figurelab_instance.generate_figure_spec("PaperChain Architecture")
paper_ir = figurelab_instance.attach_to_paper_ir(paper_ir, figure_spec)
print("\n=== Figure Attached ===")
print(paper_ir["figures"])

# 9. Export Renderers
exporters = export_renderers.ExportRenderers()
word_file = exporters.export_word(paper_ir)
tex_file = exporters.export_latex(paper_ir)
print("\n=== Export Results ===")
print("Word:", word_file)
print("LaTeX:", tex_file)
```
